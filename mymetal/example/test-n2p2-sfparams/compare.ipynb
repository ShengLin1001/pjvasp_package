## Compare with n2p2

```python
import os
import numpy as np
from mymetal.io.general import general_read
n2p2dir = "./n2p2/"
fname = os.path.join(n2p2dir, 'function.data')

# Taken from T. L.
with open(fname,'r') as f:
    counter = 0
    i_frm = 0
    sp = []
    feat = []
    sp_frm = []
    feat_frm = []
    prop_frm = []
    nat_frm = []
    for line in f:                                                   
        split = line.split()
    # number of atoms in this frame
        if len(split) == 1:
            nat = int(split[0])
            nat_frm.append(nat)
    # counting
        elif counter < nat:
            sp_frm.append(int(split[0]))  					# kind of atom
            feat_frm.append(np.array([float(s) for s in split[1:]]))		# collect features
            counter += 1
    # last line of every frame
        else:
            i_frm += 1                                                     # number of structures
            counter = 0
    
    n_frm = i_frm
    feat_frm = np.array(feat_frm)

np.savetxt('raw_n2p2_all_sf.txt', feat_frm, fmt='%16.10f')
```

```python
# Compare two array: my_sf_all and feat_frm

my_sf_all = general_read('raw_my_all_sf.txt', has_header=False)
my_sf_all = np.array(my_sf_all)

if my_sf_all.shape != feat_frm.shape:
    raise ValueError(f"Diff shape {my_sf_all.shape} {feat_frm.shape}")

match_feat_col = []
n2p2_col_list = list(range(feat_frm.shape[1]))

for col in range(my_sf_all.shape[1]):

    found = False

    for col_n2p2 in n2p2_col_list:
        # Order is different, so just loop
        # If all columns match
        if np.allclose(my_sf_all[:, col], feat_frm[:, col_n2p2], atol=1e-8):
            match_feat_col.append(col_n2p2)
            n2p2_col_list.remove(col_n2p2)
            found = True
            break
    
    # If still not found
    if not found:
        print(f'No matching column found for column {col}')
        # This is false tag

if len(match_feat_col) != len(set(match_feat_col)):
    raise ValueError(f"Duplicate columns found in match_feat_col")
```
